```
import pandas as pd
df = pd.read_csv('iris.csv')
df.info()
```

```
if True in df.isna().any().values:
    df = df.dropna()

df.info()
```

```
import numpy as np
species = np.unique(df['species'])

for ispec in species:
    print(ispec)
    print()
    print('Average')
    print(df[df['species'] == ispec].mean())
    print('Variance')
    print(df[df['species'] == ispec].var())
    print()
    print()
```

```
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.figure(figsize=(9,9))

i=1
for key in df.keys().values:
    if key != 'species':
        plt.subplot(2,2,i)
        for ispec in species:
            plt.hist(df[key][df['species']==ispec],bins=12,label=ispec,alpha=0.5)

        plt.xlabel(key,fontsize=18)
        plt.ylabel('number of records',fontsize=18)
        plt.xticks(fontsize=18)
        plt.yticks(fontsize=18)
        plt.legend(loc='upper right',fontsize=16)
        i+=1
plt.show()
```
